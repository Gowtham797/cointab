# Applicable weight for company

```python
def weight_slab(row):
    weight = row["Total weight as per X (KG)"]
    slab = row["Slab length as per zone x(KG)"]
    i = weight % slab
    if i == weight:
        return slab
    elif i > 0:
        return int(weight / slab) * slab + slab
    else:
        return weight
```

```python
df_company_slab["Applicable weight as per X (KG)"] = df_company_slab.apply(lambda row: weight_slab(row), axis=1)
```

# Expected Charge as per X(KG)

```python
df_x_charge=df_final[["Order ID","Applicable weight as per X (KG)","Delivery Zone as per X","Type of Shipment"]]
```

```python
df_x_charge=pd.merge(df_x_charge,df_courier_rates,left_on="Delivery Zone as per X",right_on="Zone",how="inner")
```

```python
df_x_charge.drop("Zone",axis=1,inplace=True)
```

```python
df_x_charge
```

```python
total_charge_list = []
for i in df_x_charge.index:
    row = df_x_charge.iloc[i]
    if row['Type of Shipment'] == "Forward charges":
        fixed_price = row['Forward Fixed Charge']
        additional_price = row['Forward Additional Weight Slab Charge']* (row['Applicable weight as per X (KG)']/row['Weight Slabs'] - 1)
        total_price = fixed_price + additional_price
        total_charge_list.append(total_price)
    elif row['Type of Shipment'] == "Forward and RTO charges":
        fixed_price = row['Forward Fixed Charge'] + row['RTO Fixed Charge']
        additional_price = (row['RTO Additional Weight Slab Charge']+row['Forward Additional Weight Slab Charge'])* (row['Applicable weight as per X (KG)']/row['Weight Slabs'] - 1)
        total_price = fixed_price + additional_price
        total_charge_list.append(total_price)
```

```python
df_x_charge['Expected Charge as per x'] = total_charge_list
df_x_charge
```

```python
df_x_charge=df_x_charge[["Order ID","Expected Charge as per x"]]
```

```python
df_final=pd.merge(df_final,df_x_charge,on="Order ID",how="inner")
```

# Summary Table 

```python
df_Summary = df_final[['Order ID','Charges Billed by Courier Company (Rs.)','Difference Between Expected Charges and Billed Charges (Rs.)']].copy()
df_Summary.rename(columns={'Difference Between Expected Charges and Billed Charges (Rs.)':'Charge Difference'}, inplace = True)
df_Summary.rename(columns={'Charges Billed by Courier Company (Rs.)':'Charges Billed'}, inplace = True)
df_Summary.sort_index(ascending = True, inplace = True)
df_Summary
```

```python
status_list = []
for i in df_Summary.index:
    row = df_Summary.iloc[i]
    if row['Charge Difference'] < 0:
        status = "Over Charged"
        status_list.append(status)
    elif row['Charge Difference'] > 0:
        status = "Under Charged"
        status_list.append(status)
    elif row['Charge Difference'] == 0:
        status = "Correctly Charged"
        status_list.append(status)
        
df_Summary.loc[:,'Status'] = status_list
df_Summary
```

# Output Data 2

```python
df_Summary_table = df_Summary.groupby(['Status'])['Charge Difference'].agg(['count','sum']).reset_index()
correctly_charged_amount = df_Summary[df_Summary['Status'] == "Correctly Charged"].groupby('Status')['Charges Billed'].sum().reset_index(name = 'Amount')
df_Summary_table.loc[df_Summary_table['Status'] == "Correctly Charged",'sum'] = correctly_charged_amount['Amount']
df_Summary_table.rename(columns = {'sum': 'Amount(Rs.)','count':'Count'}, inplace= True)
df_Summary_table['Status'] = "Total orders where X has been " + df_Summary_table['Status']
df_Summary_table
```
